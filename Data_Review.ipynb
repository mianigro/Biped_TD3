#### Importing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = pd.read_csv("log_file.csv", index_col="Epoch", header=None, names=["Epoch", "Score", "Tau"])
data_file[:5]
```

#### Line Plot - Raw Score

```python
plt.plot(data_file["Score"])
plt.show()
```

#### Line Plot - 20 Episode Rolling Average

```python
rolling_ave = data_file["Score"].rolling(window=20).mean()
```

```python
plt.plot(rolling_ave)
plt.show()
```
